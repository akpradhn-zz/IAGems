#### Image transform using Furier Transform

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

#### Frequency Transform of German Shephard

```python
img = cv2.imread('german.jpg',0)
print('Dimenion of the Image is :',img.shape)
```

```python
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20*np.log(np.abs(fshift))
```

```python
plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum, cmap ='gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

The whiter region is the low frequency region.

```python
#Defining Image Shape
rows, cols = img.shape
nrow,ncol = np.int(rows/2) , np.int(cols/2)

"""
# Removing low frequency
From the 201 x 251 image size we subset the central potion by taking
a dimention of 60x60 to remove the low frequencies.
"""
fshift[nrow-30:nrow+30, ncol-30:ncol+30] = 0
```

##### Apply inverse shift 

```python
f_ishift = np.fft.ifftshift(fshift)
img_back = np.fft.ifft2(f_ishift)
img_back = np.abs(img_back)

plt.subplot(131),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(img_back, cmap = 'gray')
plt.title('Image after HPF'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(img_back)
plt.title('Result in JET'), plt.xticks([]), plt.yticks([])

plt.show()
```
